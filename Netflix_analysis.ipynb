```python
import pandas as pd
df = pd.read_csv("C:/Users/Akash/Downloads/netflix_titles.csv/netflix_titles.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
```

Dataset Description:
- show_id: Unique ID
- type: Movie or TV Show
- title: Name of content
- country: Country of production
- release_year: Year released
- rating: Audience rating
- listed_in: Genre


```python
df = df.dropna(subset=["date_added"])
df.isnull().sum()
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df["rating"] = df["rating"].fillna("Not Rated")
```

```python
df.isnull().sum()
```

Data Cleaning:
- Removed rows with missing date_added.
- Filled missing country values with 'Unknown'.
- Filled missing rating values with 'Not Rated'.
- Director and cast columns were left unchanged.


```python
df["type"].value_counts()
```

```python
df["release_year"].value_counts().head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(y = "release_year" , data = df , order = df["release_year"].value_counts().head(10).index)
plt.title("Top 10 Release Years on Netflix")
plt.show()
```

```python
df["rating"].value_counts()
```

```python
df["country"].value_counts().head(10)
```

Insights:
1. Movies are more common than TV Shows on Netflix.
2. Content production increased significantly after 2015.
3. TV-MA and TV-14 are the most frequent ratings.
4. United States contributes the highest number of titles.


```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.countplot(x = "type" , data = df)
plt.title("Movies vs TV shows on Netflix")
plt.show()
```

```python
df_year = df["release_year"].value_counts().sort_index()

sns.lineplot(x = df_year.index, y = df_year.values)
plt.title("Netflix content growth over years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
df_countries = df["country"].value_counts().head(10)
sns.barplot(x =  df_countries.values , y = df_countries.index)
plt.title("Top 10 countries with most Netflix content")
plt.show()
```

Insights:
1. Netflix has more movies than TV shows.
2. Content production has grown rapidly after 2015.
3. The United States dominates Netflix content production.
4. Visualizations help clearly communicate trends and comparisons.
